```python
dataset = pd.read_csv('insurance_pre.csv')
```

```python
dataset = pd.get_dummies(dataset,drop_first=True).astype(int)
dataset.head(3)
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent = dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.3,random_state = 42)
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
```

```python
ada_regressor1 = AdaBoostRegressor(
    estimator=RandomForestRegressor(max_depth=3),
    n_estimators=100,
    learning_rate=0.5,
    random_state=42
)
ada_regressor1.fit(X_train,y_train)
```

```python
ada_regressor2 = AdaBoostRegressor(
    estimator=DecisionTreeRegressor(max_depth=3),
    n_estimators=100,
    learning_rate=0.5,
    random_state=42
)
ada_regressor2.fit(X_train,y_train)
```

```python
ada_regressor3 = AdaBoostRegressor(
    estimator=SVR(),
    n_estimators=100,
    learning_rate=0.5,
    random_state=42
)
ada_regressor3.fit(X_train,y_train)
```

```python
y_pred1 = ada_regressor1.predict(X_test)
y_pred2 = ada_regressor2.predict(X_test)
y_pred3 = ada_regressor3.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_square1 = r2_score(y_test,y_pred1)
r_square1
```

```python
r_square2 = r2_score(y_test,y_pred2)
r_square2
```

```python
r_square3 = r2_score(y_test,y_pred3)
r_square3
```
